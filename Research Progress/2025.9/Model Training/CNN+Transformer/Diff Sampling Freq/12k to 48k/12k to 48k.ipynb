# 48k降采样至12k

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import loadmat
from scipy.signal import resample
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 中文
matplotlib.rcParams['axes.unicode_minus'] = False   # 解决负号显示问题

# ===============================
# 1. 文件路径配置（12kDE用于训练/验证，48kDE用于测试）
# ===============================
file_paths_12k = [
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.007/B007_0.mat",   # B007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.014/B014_0.mat",   # B014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.021/B021_0.mat",   # B021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.007/IR007_0.mat",  # IR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.014/IR014_0.mat",  # IR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.021/IR021_0.mat",  # IR021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.007/OR007@6_0.mat",# OR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.014/OR014@6_0.mat",# OR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.021/OR021@6_0.mat",# OR021
    
]
file_paths_48k = [
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.007/B007_0.mat",   # B007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.014/B014_0.mat",   # B014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.021/B021_0.mat",   # B021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.007/IR007_0.mat",  # IR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.014/IR014_0.mat",  # IR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.021/IR021_0.mat",  # IR021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.007/OR007@6_0.mat",# OR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.014/OR014@6_0.mat",# OR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.021/OR021@6_0.mat",# OR021
    
]
labels = list(range(9))

segment_length = 256
val_size  = 0.2
overlap   = 0.0

save_dir = r"D:\study\code\Model Training\CNN+Transformer\Diff Sampling Freq\12k to 48k"

# ===============================
# 2. 数据预处理函数
# ===============================
def load_and_segment_data(file_paths, labels, segment_length, overlap, downsample_ratio=1):
    """downsample_ratio: e.g. 4 表示把48k降采样到12k"""
    X, y = [], []
    step = int(segment_length * (1 - overlap))  # 步长
    for file, label in zip(file_paths, labels):
        mat_dict = loadmat(file)
        key = [k for k in mat_dict.keys() if "DE_time" in k][0]
        signal = mat_dict[key].squeeze()

        # 如果需要降采样
        if downsample_ratio > 1:
            new_len = len(signal) // downsample_ratio
            signal = resample(signal, new_len)

        # 切片
        segments = []
        for start in range(0, len(signal) - segment_length + 1, step):
            seg = signal[start:start + segment_length]
            segments.append(seg)
        segments = np.array(segments)

        X.append(segments)
        y.extend([label] * len(segments))

    X = np.vstack(X)
    y = np.array(y)
    return X, y

# ===============================
# 3. 数据加载
# ===============================
save_files = [os.path.join(save_dir, f) for f in [
    "X_train.npy","y_train.npy","X_val.npy","y_val.npy","X_test.npy","y_test.npy"
]]

if all(os.path.exists(f) for f in save_files):
    print("✅ 发现已有预处理数据，直接加载")
    X_train = np.load(os.path.join(save_dir, "X_train.npy"))
    y_train = np.load(os.path.join(save_dir, "y_train.npy"))
    X_val   = np.load(os.path.join(save_dir, "X_val.npy"))
    y_val   = np.load(os.path.join(save_dir, "y_val.npy"))
    X_test  = np.load(os.path.join(save_dir, "X_test.npy"))
    y_test  = np.load(os.path.join(save_dir, "y_test.npy"))
else:
    print("⚙️ 重新从 .mat 文件生成数据")
    X12k, y12k = load_and_segment_data(file_paths_12k, labels, segment_length, overlap, downsample_ratio=1)
    X48k, y48k = load_and_segment_data(file_paths_48k, labels, segment_length, overlap, downsample_ratio=4)

    # 训练集/验证集（12k）
    X_train_full, X_val, y_train_full, y_val = train_test_split(
        X12k, y12k, test_size=val_size, stratify=y12k, random_state=42
    )

    # 归一化
    mean, std = np.mean(X_train_full), np.std(X_train_full)
    X_train = ((X_train_full - mean) / std)[..., np.newaxis]
    X_val   = ((X_val       - mean) / std)[..., np.newaxis]
    X_test  = ((X48k        - mean) / std)[..., np.newaxis]

    y_train = y_train_full
    y_test  = y48k

    os.makedirs(save_dir, exist_ok=True)
    np.save(os.path.join(save_dir, "X_train.npy"), X_train)
    np.save(os.path.join(save_dir, "y_train.npy"), y_train)
    np.save(os.path.join(save_dir, "X_val.npy"), X_val)
    np.save(os.path.join(save_dir, "y_val.npy"), y_val)
    np.save(os.path.join(save_dir, "X_test.npy"), X_test)
    np.save(os.path.join(save_dir, "y_test.npy"), y_test)

print("训练集大小:", X_train.shape, y_train.shape)
print("验证集大小:", X_val.shape,   y_val.shape)
print("跨采样率测试集大小(48k→12k):", X_test.shape, y_test.shape)

# ===============================
# 4. CNN+Transformer 模型
# ===============================
def transformer_block(x, head_size, num_heads, ff_dim, dropout=0):
    x_norm = layers.LayerNormalization(epsilon=1e-6)(x)
    attn_output = layers.MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(x_norm, x_norm)
    x = layers.Add()([x, attn_output])

    x_norm = layers.LayerNormalization(epsilon=1e-6)(x)
    ff_output = layers.Dense(ff_dim, activation="relu")(x_norm)
    ff_output = layers.Dense(x.shape[-1])(ff_output)
    x = layers.Add()([x, ff_output])
    return x

def build_cnn_transformer(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(32, 7, padding="same", activation="relu")(inputs)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Conv1D(64, 5, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(2)(x)
    for _ in range(2):
        x = transformer_block(x, head_size=32, num_heads=2, ff_dim=64, dropout=0.1)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(0.2)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    return models.Model(inputs, outputs)

num_classes = len(np.unique(y_train))
model = build_cnn_transformer(input_shape=(segment_length,1), num_classes=num_classes)
model.summary()

# ===============================
# 5. 训练模型
# ===============================
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

callbacks = [
    ReduceLROnPlateau(factor=0.5, patience=5, verbose=1),
    EarlyStopping(patience=10, restore_best_weights=True, verbose=1)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=30,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# ===============================
# 6. 结果评估
# ===============================
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"跨采样率测试集准确率: {test_acc:.4f}")

plt.figure(figsize=(10,4))
plt.plot(history.history["accuracy"], label="训练集")
plt.plot(history.history["val_accuracy"], label="验证集")
plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend(); plt.title("学习曲线")
plt.show()

y_pred = np.argmax(model.predict(X_test), axis=1)
cm = confusion_matrix(y_test, y_pred)
class_names = ["B007","B014","B021","IR007","IR014","IR021","OR007","OR014","OR021"]

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel("预测值"); plt.ylabel("真实值"); plt.title("混淆矩阵")
plt.show()

print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import loadmat
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 或 'Microsoft YaHei'
matplotlib.rcParams['axes.unicode_minus'] = False   # 解决负号显示问题

# ===============================
# 1. 文件路径配置（12k -> train/val，48k -> test）
# ===============================
file_paths_12k = [
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.007/B007_0.mat",   # B007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.014/B014_0.mat",   # B014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.021/B021_0.mat",   # B021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.007/IR007_0.mat",  # IR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.014/IR014_0.mat",  # IR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.021/IR021_0.mat",  # IR021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.007/OR007@6_0.mat",# OR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.014/OR014@6_0.mat",# OR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/12k Drive End Bearing Fault Data/0.021/OR021@6_0.mat",# OR021
]

file_paths_48k = [
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.007/B007_0.mat",   # B007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.014/B014_0.mat",   # B014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.021/B021_0.mat",   # B021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.007/IR007_0.mat",  # IR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.014/IR014_0.mat",  # IR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.021/IR021_0.mat",  # IR021
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.007/OR007@6_0.mat",# OR007
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.014/OR014@6_0.mat",# OR014
    r"D:/HuaweiMoveData/Users/hps_h/Desktop/CWRU/48k Drive End Bearing Fault Data/0.021/OR021@6_0.mat",# OR021
]

labels = list(range(9))   # 9 类（不含 Normal）

segment_length_train = 256
segment_length_test  = 1024
val_size  = 0.2
overlap   = 0.0

save_dir = r"D:\study\code\Model Training\CNN+Transformer\Diff Sampling Freq\12k to 48k"

# ===============================
# 2. 数据预处理函数
# ===============================
def load_and_segment_data(file_paths, labels, segment_length, overlap):
    X, y = [], []
    step = int(segment_length * (1 - overlap))
    for file, label in zip(file_paths, labels):
        mat_dict = loadmat(file)
        key = [k for k in mat_dict.keys() if "DE_time" in k][0]
        signal = mat_dict[key].squeeze()

        segments = []
        for start in range(0, len(signal) - segment_length + 1, step):
            seg = signal[start:start + segment_length]
            segments.append(seg)
        segments = np.array(segments)

        X.append(segments)
        y.extend([label] * len(segments))

    X = np.vstack(X)
    y = np.array(y)
    return X, y

# ===============================
# 3. 数据加载/生成
# ===============================
save_files = [os.path.join(save_dir, f) for f in ["X_train.npy","y_train.npy","X_val.npy","y_val.npy","X_test.npy","y_test.npy"]]

if all(os.path.exists(f) for f in save_files):
    print("✅ 已有数据，直接加载")
    X_train = np.load(os.path.join(save_dir, "X_train.npy"))
    y_train = np.load(os.path.join(save_dir, "y_train.npy"))
    X_val   = np.load(os.path.join(save_dir, "X_val.npy"))
    y_val   = np.load(os.path.join(save_dir, "y_val.npy"))
    X_test  = np.load(os.path.join(save_dir, "X_test.npy"))
    y_test  = np.load(os.path.join(save_dir, "y_test.npy"))
else:
    print("⚙️ 重新生成数据")
    X12, y12 = load_and_segment_data(file_paths_12k, labels, segment_length_train, overlap)
    X48, y48 = load_and_segment_data(file_paths_48k, labels, segment_length_test, overlap)

    X_train_full, X_val, y_train_full, y_val = train_test_split(
        X12, y12, test_size=val_size, stratify=y12, random_state=42
    )

    mean, std = np.mean(X_train_full), np.std(X_train_full)
    X_train = ((X_train_full - mean) / std)[..., np.newaxis]
    X_val   = ((X_val       - mean) / std)[..., np.newaxis]
    X_test  = ((X48         - mean) / std)[..., np.newaxis]   # 归一化用 12k 统计量

    y_train, y_test = y_train_full, y48

    os.makedirs(save_dir, exist_ok=True)
    np.save(os.path.join(save_dir, "X_train.npy"), X_train)
    np.save(os.path.join(save_dir, "y_train.npy"), y_train)
    np.save(os.path.join(save_dir, "X_val.npy"), X_val)
    np.save(os.path.join(save_dir, "y_val.npy"), y_val)
    np.save(os.path.join(save_dir, "X_test.npy"), X_test)
    np.save(os.path.join(save_dir, "y_test.npy"), y_test)

print("训练集大小:", X_train.shape, y_train.shape)
print("验证集大小:", X_val.shape,   y_val.shape)
print("跨采样测试集大小(48k):", X_test.shape, y_test.shape)

# ===============================
# 4. CNN+Transformer 模型
# ===============================
def transformer_block(x, head_size, num_heads, ff_dim, dropout=0):
    x_norm = layers.LayerNormalization(epsilon=1e-6)(x)
    attn_output = layers.MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(x_norm, x_norm)
    x = layers.Add()([x, attn_output])

    x_norm = layers.LayerNormalization(epsilon=1e-6)(x)
    ff_output = layers.Dense(ff_dim, activation="relu")(x_norm)
    ff_output = layers.Dense(x.shape[-1])(ff_output)
    x = layers.Add()([x, ff_output])
    return x

def build_cnn_transformer(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    x = layers.Conv1D(32, 7, padding="same", activation="relu")(inputs)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Conv1D(64, 5, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(2)(x)

    for _ in range(2):
        x = transformer_block(x, head_size=32, num_heads=2, ff_dim=64, dropout=0.1)

    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(0.2)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return models.Model(inputs, outputs)

num_classes = len(np.unique(y_train))

# 🚨 改这里：输入 shape 用 (None,1)，允许变长
model = build_cnn_transformer(input_shape=(None,1), num_classes=num_classes)
model.summary()

# ===============================
# 5. 训练
# ===============================
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

callbacks = [
    ReduceLROnPlateau(factor=0.5, patience=5, verbose=1),
    EarlyStopping(patience=10, restore_best_weights=True, verbose=1)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=30,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# ===============================
# 6. 评估
# ===============================
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"测试集准确率: {test_acc:.4f}")

plt.figure(figsize=(10,4))
plt.plot(history.history["accuracy"], label="训练集")
plt.plot(history.history["val_accuracy"], label="验证集")
plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend(); plt.title("学习曲线")
plt.show()

y_pred = np.argmax(model.predict(X_test), axis=1)
cm = confusion_matrix(y_test, y_pred)
class_names = ["B007","B014","B021","IR007","IR014","IR021","OR007","OR014","OR021"]

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel("预测值"); plt.ylabel("真实值"); plt.title("混淆矩阵")
plt.show()

print(classification_report(y_test, y_pred, target_names=class_names))
```
